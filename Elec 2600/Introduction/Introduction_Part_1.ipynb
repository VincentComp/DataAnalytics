### Include necessary libraries

```python
import numpy as np
import matplotlib.pyplot as plt

import pylab as pl
```

### Ex. 01 - Generate array and Use for loop

```python
a = np.linspace(0, 1, 5)   # start, stop, number of points
b = np.arange(0, 1, 2e-1)  # start, (stop - step), step = 0.2
c = np.arange(1,5)         # step = 1 by default

print('a array = ', a)
print('b array = ', b)
print('c array = ', c)
print(' ')

for i in range(5):   # index i = 0,1,2,3,4
    print(f'a[{i}] = {a[i]}')
```

### Ex. 02 - Generate a plot

```python
t = np.arange(0, 2, 5e-2)  # start, (stop - step), step = 0.05
x1 = np.cos(2*pl.pi*t)
x2 = np.sin(2*pl.pi*t)

fig, ax = plt.subplots()

ax.plot(t, x1, 'ro', label = 'cosine')  
ax.plot(t, x2, 'b--', label = 'sine')

plt.grid()
plt.xlabel('time (in seconds)')
plt.ylabel('x1(t) and x2(t)')

plt.vlines(0, -1.2, 1.2, 'g') #add lines
plt.hlines(0, -0.2, 2.2, 'g')

plt.xticks(np.arange(0, 2.2, 0.2))    #add interval
plt.yticks(np.arange(-1.2, 1.2, 0.2))

plt.legend(loc = [1, 0.8])
plt.title('Generate a plot')

plt.show()
```

### Ex. 03 - Addition, Subtraction, Multiplication and division

```python
a = np.array([2, 4.2, 6.6, 8.8, 10.5])
b = np.array([1, 2, 3, 4, 5])

print('a array = ', a)
print('b array = ', b)
print(' ')

print('a + b = ', a + b)
print('a - b = ', a - b)
print('a * b = ', a * b)
print('a / b = ', a / b)
print(' ')
print('a // b = ', a // b)
print(' ')

print('a + b = ', np.add(a, b))
print('a - b = ', np.subtract(a, b))
print('a * b = ', np.multiply(a, b))
print('a / b = ', np.divide(a, b))
print(' ')
print(' ')

c = [2, 4.2, 6.6, 8.8, 10.5]
d = [1, 2, 3, 4, 5]          

print('c array = ', c)
print('d array = ', d)
print(' ')

print('c + d = ', c + d)  # equivalent to Append(bc it's list)
print(' ')

print('c + d = ', np.add(c, d))
print('c - d = ', np.subtract(c, d))
print('c * d = ', np.multiply(c, d))
print('c / d = ', np.divide(c, d))
```

### Ex. 04 - Append, Power, Square root

```python
a = np.array([9, 16, 25, 36, 49])
b = np.array([1, 2, 3, 4, 5])

print('Combine arrays a and b = ', np.append(a, b))
print(' ')

c = np.power(b, 2)
print('Power of a = ', c)

d = np.sqrt(a)
print('Power of a = ', d)
```

### Ex. 05 - Size, Length, Dimension and Sum

```python
a = np.array( [(1, 3, 5, 7, 9), (2, 4, 6, 8, 10)] )  # 2-by-2 array

print('a array = \n', a)
print(' ')

print('The number of elements = ', a.size)
print('The length of array = ', len(a))
print('The dimension of array = ', a.ndim)
print(' ')

print('Sum all elements = ', np.sum(a))
print(' ')

print('Element-wise sum = ', sum(a))   # if array is more than one length
print(' ')
print(' ')

b = np.array([1, 2, 3, 4, 5])
print('b array = ', b)
print(' ')

print('Sum all elements = ', np.sum(b))
print(' ')
print('Sum all elements = ', sum(b))
```

### Ex. 06 - Equal and Cumsum

```python
a = np.array([1, 4, 5])
b = np.array([1, 4, 5, 5, 4, 1, 1, 1])

for i in range(a.size):
    print(f'a{[i]} = ', a[i])
    compare = np.equal(b, a[i])#return array
    print(compare)
    count = np.cumsum(compare) #return array of cumulative true
    print(count)
    print(' ')
```
